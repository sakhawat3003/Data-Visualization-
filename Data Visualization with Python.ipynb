```python
#creating data for plotting
years= [2013 + x for x in range(10)]
salary= np.random.randint(55,110,10)
salary=np.sort(salary)
```

```python
print(years)
print(salary)
```

```python
#customization with labels and titles
plt.plot(years, salary)
plt.title('Yearly Salary of Sakaboo (in USD)')
plt.xlabel('Year')
plt.ylabel('Salary in USD')
#we need to create y ticks for y-axis within the range of the salaries
y_ticks=list(range(50,115,5))
plt.yticks(y_ticks);
```

```python
#the y-axis should be in thousand times of the values. We can add 'k' after each number in the y-axis.
plt.plot(years, salary)
plt.title('Yearly Salary of Sakaboo (in USD)')
plt.xlabel('Year')
plt.ylabel('Salary in USD')
plt.yticks(y_ticks, [f"{x}k" for x in y_ticks]); #first it places the y ticks according to the number then it follows the
#names from the list
```

```python
#creating a couple of random stock price data for line plot
stock_1=np.random.randint(80,110,10)
stock_2=np.random.randint(85,115,10)
stock_3=np.random.randint(83,113,10)
```

```python
#line plot. When we provide only one argument in the plot function the x-axis is created automatically.
plt.plot(stock_1)
plt.plot(stock_2)
plt.plot(stock_3);
```

```python
#but there is a problem. We can't distinguish between the different plots belonging to which stock. 
#In that case, we need to provide legend for clarification. 
plt.plot(stock_1, label='stock one')
plt.plot(stock_2, label='stock two')
plt.plot(stock_3, label='stock three')
plt.legend();
```

```python
#we can also plot multiple plots at once. 
x=np.arange(1,100) #a sequence of number for x-axis from 1 to 100 with intervals of 1
figure, axis = plt.subplots(2,2) #creating places for 4 plots in a 2*2 grid
axis[0,0].plot(x, np.sin(x)) #plotting a sine wave function in the first row and first column
axis[0,0].set_title('Sine Wave')

axis[0,1].plot(x, np.random.random(99)) #plotting a random number function in the first row and second column
axis[0,1].set_title('Random Function')

axis[1,0].plot(x, np.cos(x)) #plotting a cosine wave function in the second row and first column
axis[1,0].set_title('Cosine Wave')

axis[1,1].plot(x, np.log(x)) #plotting a log function in the second row and second column
axis[1,1].set_title('Logarithmic Function');
```

```python
#3-D plot
x=np.arange(-5,5,0.1)
y=np.arange(-5,5,0.1)
z=x**2+y**2
```

```python
#this will show a simple line plot but in a 3d space
ax=plt.axes(projection='3d')
ax.plot(x,y,z)
ax.set_title('3-D Plot')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis');
```

```python
#now we will create a surface plot by all the possible combinations of x and y values
X,Y=np.meshgrid(x,y)
Z=np.sin(X)*np.cos(Y)
```

```python
from mpl_toolkits import mplot3d #this is necessary for plotting 3D surface in Jupyter notebook
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot_surface(X,Y,Z);
```

```python
#Lets add a colormap to the surface plot
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot_surface(X,Y,Z, cmap='Spectral');
```
